```python
import altair as alt
from vega_datasets import data
```

```python

source = data.windvectors()
```

```python
source.head()
```

```python
alt.Chart(source).mark_point(shape="wedge", filled=True).encode(
    latitude="latitude",
    longitude="longitude",
    color=alt.Color(
        "dir", scale=alt.Scale(domain=[0, 360], scheme="rainbow"), legend=None
    ),
    angle=alt.Angle("dir", scale=alt.Scale(domain=[0, 360], range=[180, 540])),
    size=alt.Size("speed", scale=alt.Scale(rangeMax=500)),
).project("equalEarth")
```

```python
alt.Chart(source).mark_point(shape="wedge", filled=False).encode(
    latitude="latitude",
    longitude="longitude",
    color=alt.Color(
        "dir", scale=alt.Scale(domain=[0, 360], scheme="rainbow")
    ),
    angle=alt.Angle("dir", scale=alt.Scale(domain=[0, 360], range=[180, 540])),
    size=alt.Size("speed", scale=alt.Scale(rangeMax=500)),
).project("equalEarth")
```

## Ejemplo grafica de barras

```python
import altair as alt
import pandas as pd
from vega_datasets import data as vega_data
```

```python
data=vega_data.gapminder()
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data_mex=data[data.country == 'Mexico']
```

```python
data_mex.head()
```

```python
alt.Chart(data_mex).mark_bar().encode(
    x='year:Q',#cuantitativo
    y='pop',
    color=alt.condition(
        alt.datum.year == 2000,  # If the year is 2000 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('chartreuse')   # And if it's not true it sets the bar steelblue.
    )
)
```

```python
alt.Chart(data_mex).mark_bar().encode(
    x='year:O',#Ordinal
    y='pop:Q',
    color=alt.condition(
        alt.datum.year == 2000,  # If the year is 2000 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('chartreuse')   # And if it's not true it sets the bar steelblue.
    )
)
```

```python
alt.Chart(data_mex).mark_bar().encode(
    x=alt.X('year:O',title=''),#Ordinal
    y=alt.Y('pop:Q',title='Poblacin'),
    color=alt.condition(
        alt.datum.year == 2000,  # If the year is 2000 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('chartreuse')   # And if it's not true it sets the bar steelblue.
    )
)
```

```python
alt.Chart(data_mex).mark_bar().encode(
    x=alt.X('year:O',title='',axis=alt.Axis(labelAngle=0,labelFontSize=11)),#Ordinal
    y=alt.Y('pop:Q',title='Poblacin',axis=alt.Axis(format='s')),
    color=alt.condition(
        alt.datum.year == 2000,  # If the year is 2000 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('chartreuse')   # And if it's not true it sets the bar steelblue.
    )
).properties(
    title=alt.TitleParams('Poblacion de Mexico de 1955 a 2005',subtitle='25 de enero del 2021',fontSize=22),
    width=600)
```

```python
alt.Chart(data_mex).mark_bar(color='red',stroke='black').encode(
    x=alt.X('year:O',title='',axis=alt.Axis(labelAngle=0,labelFontSize=11)),#Ordinal
    y=alt.Y('pop:Q',title='Poblacin',axis=alt.Axis(format='s'))
).properties(
    title=alt.TitleParams('Poblacion de Mexico de 1955 a 2005',subtitle='25 de enero del 2021',fontSize=22),
    width=600)
```
